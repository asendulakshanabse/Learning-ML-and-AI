```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Iris.csv")
x = data.drop(["Species", "Id"], axis=1)
y = data["Species"]
```

```python
x.head()
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC(C=1, kernel='rbf')
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
# gridSearchCV
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10],
              'kernel': ['rbf', 'linear', 'poly']
            }           
```

```python
grid_search = GridSearchCV(estimator=model, param_grid=param_grid)
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.score(x_test, y_test)
```

```python
# random search
from sklearn.model_selection import RandomizedSearchCV
param_dist = {'C': [0.1, 1, 10],
              'kernel': ['rbf', 'linear', 'poly']}
```

```python
randomize_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=5)
randomize_search.fit(x_train, y_train)
```

```python
randomize_search.best_params_
```

```python
randomize_search.score(x_test, y_test)
```
